```python
import numpy as np
import pandas as pd
import zipfile

zip_file_path = '/content/drug+review+dataset+druglib+com.zip'

file_within_zip = 'drugLibTrain_raw.tsv'

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    with zip_ref.open(file_within_zip) as file:
        df = pd.read_csv(file, sep='\t')
```

**Chart 1** - **Bar chart**

```python

df['rating'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

**Chart 2** - **Pie Chart**

```python
df['condition'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(12, 12))
plt.title('Distribution of Conditions')
plt.ylabel('')
plt.show()
```

**Chart 3** - **Scatter Plot**

```python
df.plot.scatter(x='rating', y='effectiveness', color='purple', alpha=0.5)
plt.title('Rating vs Effectiveness')
plt.xlabel('Rating')
plt.ylabel('Effectiveness')
plt.show()
```
